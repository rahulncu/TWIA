### 1. Import a file and find the frequency of each word in a file.

```python
from nltk.corpus import shakespeare 
```

```python
shakespeare.fileids()
```

```python
# import file from shakespeare
shk_file=shakespeare.xml('dream.xml')
```

```python
text=[]
for p in shk_file:
    text.extend(list(p.itertext()))
    
```

```python
text=" ".join(text)
```

```python
regex_token = nltk.RegexpTokenizer(r"\w+")
```

```python
wordtoken=regex_token.tokenize(text)
```

```python
from nltk.probability import FreqDist
```

```python
wordfreq= dict(FreqDist(wordtoken))
```

```python
print("Frequency of each word in the file : \n")
wordfreq
```

### 2. Find the five most frequently occurring words from the document.

```python
wordfreq_key=list(wordfreq.keys())
wordfreq_value=list(wordfreq.values())
```

```python
top5word_freq=sorted(wordfreq_value,reverse=True)[0:5]
```

```python
top5word=[]
for i in top5word_freq:
    top5word.append(wordfreq_key[wordfreq_value.index(i)])
```

```python
print('Five most frequently occuring words : ',top5word)
```

### 3. Import wordnet corpus from the available nltk corpus list and find out the synset of word “bank”. Also list the definition and example of first sysnst in the list.

```python
from nltk.corpus import wordnet
```

```python
synset_list=[]
for word in wordnet.synsets('bank'):
    synset_list.append(word)
```

```python
synset_list
```

```python
print('Definition : ',synset_list[0].definition(),'\nExample : ',synset_list[0].examples())
```
